```python
import PyPDF2 
import re
```

```python
def dateformat(str):
    for strngs in str:
        a=re.findall(r'(^([0-2][0-9]|3[0-1])\/(0[0-9]|1[0-2])\/\d{4}$)', strngs) # 'dd/mm/yy'
        b=re.findall(r'(^([0-2][0-9]|3[0-1])\.(0[0-9]|1[0-2])\.\d{4}$)', strngs) # 'dd.mm.yyyy'
        c=re.findall(r'(^\d{4}-(0[0-9]|1[0-2])-([0-2][0-9]|3[0-1])$)', strngs)     # 'yyyy-mm-dd'
        d=re.findall(r'^[January|Febuary|March|April|May|June|July|August|September|October|November|December]* ([0-2][0-9]|(3)[0-1]), \d{4}$'
                     , strngs) # 'August 21, 1999'
        e=re.findall(r'^([0-2][0-9]|(3)[0-1]) (Nov|Aug)(,) (\d{4})$' , strngs) # '21 Aug, 1999'
        if(len(a)>0):
            print(strngs,':dd/mm/yy')
        if(len(b)>0):
            print(strngs,':dd.mm.yyyy')
        if(len(c)>0):
            print(c,':yyyy-mm-dd')
        if(len(d)>0):
            print(strngs,':August 21, 1999')
        if(len(e)>0):
            print(strngs,':21 Aug, 1999')
```

```python
def timeformat(str):
    for strngs in str:
        a=re.findall(r'(^\d{0,11}\.\d{0,59} [A|P]M$)',strngs) # '4.30 AM'
        b=re.findall(r'(^\d{0,11}:\d{0,59} [a|p]m$)',strngs) # '4:30 am'
        c=re.findall(r'(^\d{0,11}\.\d{0,59} [a|p]m$)',strngs) # '5.15 pm'
        d=re.findall(r'(^\d{0,24}:\d{0,59}$)',strngs) # '13:24'
        if(len(a)>0):
            print(strngs,':H.MM AM')
        if(len(b)>0):
            print(strngs,':H:MM am')
        if(len(c)>0):
            print(strngs,':H.MM pm')
        if(len(d)>0):
            print(strngs,':HH:MM')
      
```

```python
def Mobileformat(str):
    for strngs in str:
        a=re.findall(r'^\d{10}$',strngs) # '7382293451'
        b=re.findall(r'^\+\d{12}$',strngs) # '+917382293451'
        c=re.findall(r'^\(\+\d{2}\) ?\d{10}$',strngs) # '(+91) 7382293451'
        if(len(a)>0):
            print(strngs,':10 digit Mobile No.')
        if(len(b)>0):
            print(strngs,':Mobile No. with country Code')
        if(len(c)>0):
            print(strngs,':Country code in Brackets')
     
```

```python
def Landlineformat(str):
    for strngs in str:
        a=re.findall(r'^\d{5}$',strngs) # '22469'
        b=re.findall(r'^\d{8}$',strngs) # '04322469'
        c= re.findall(r'^\(\d\)\d{5}$',strngs) # '(043)22469'
        if(len(a)>0):
            print(strngs,':5 digit Landline No.')
        if(len(b)>0):
            print(strngs,':Landline No. with local Code')
        if(len(c)>0):
            print(strngs,':Local code in Brackets')
```

```python
def Thousandsformat(str):
    for strngs in str:
        a=re.sub(r'\d+', lambda m: format(int(m.group(0)), ','), strngs)
        print(a)
```

```python

def Currformat(str):
    for strng in str:
        Currformat1= re.findall(r'^\d+\.\d+? dollars$',strng)
        Currformat2= re.findall(r'^\$ \d+\.\d+?$',strng)
        if(len(Currformat1)>0):
            print(Currformat1)
        if(len(Currformat2)>0):
            print(Currformat2)
```

```python
def Emailformat(str):
    for strng in str:
        Emailformat1=re.findall('[\w._%+-\.].*@\w.*\.[a-zA-Z]{2,4}.+',strng)
        Emailformat2=re.findall('^[\w_%+-\.]+\[at\]\w.*\[dot\][a-zA-Z]{2,4}.+$',strng)
        Emailformat3=re.findall('^[\w_%+-\.]+\[AT\]\w.*\[DOT\][a-zA-Z]{2,4}.+$',strng)
        if(len(Emailformat1)>0):
            print(Emailformat1)
        if(len(Emailformat2)>0):
            print(Emailformat2)
        if(len(Emailformat3)>0):
            print(Emailformat3)
```

```python
def EURLformat(str):
    for strng in str:
        Urlformat1=re.findall('(http(s)?://)?([\w-]+\.)+[\w-]+[.com]+(/[/?%&=]*)?',strng)
        if(len(Urlformat1)>0):
            print(strng)
```

```python
pdfFileObj = open('WSoutput.pdf', 'rb') 
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
for i in range(0,pdfReader.numPages):
    pageObj = pdfReader.getPage(i)
    str=pageObj.extractText().split('\n')
    dateformat(str)
    timeformat(str)
    Mobileformat(str)
    Landlineformat(str)
    Currformat(str)
    Emailformat(str)
    EURLformat(str)
    
pdfFileObj.close() 
```
